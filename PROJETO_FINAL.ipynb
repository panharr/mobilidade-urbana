```python
from colour import Color

y = np.array(df['motivo_viagem'].value_counts())
x = np.array(["Trabalho", "Outros", "Estudo", "Saúde", "Lazer"])

blue = Color("blue")
colors = list(blue.range_to(Color("yellow"),10))
colors = [color.rgb for color in colors]

for index, value in enumerate(y):
    plt.text(value, index, str(value))

plt.title("Motivo da viagem")
plt.barh(x, y, color=colors, edgecolor='black')
plt.show()
```
